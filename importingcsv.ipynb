```
import pandas as pd
```

```
states_data = pd.read_csv('states_data.tsv',sep='\t')
states_data.head()
```

```
states_data.dtypes
states_data['Population'] = states_data['Population*']

states_data = states_data.drop('Population*',axis=1)


#population is an object as we see in this datatype list. so we need it to be a number. it gets pulled in a an object
#bc it has commas and pandas can't make a numbeer out of it. now we are changing the column name to just Population


#code for taking the commas out
states_data['Population'] = states_data['Population'].str.replace(',','').astype('float32')
```

```
states_data.head()
```

```
results = states_data.query('Population > 5000000')
results
```
